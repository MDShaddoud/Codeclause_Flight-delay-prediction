```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np
import warnings
from sklearn.exceptions import DataConversionWarning
```

```python
train_x, test_x, train_y, test_y = train_test_split(X,Y,test_size=0.18, random_state=42)
```

```python
rfr_model = RandomForestRegressor(n_estimators=200, random_state=42)
```

```python
rfr_model.fit(train_x, train_y)
```

```python
pred_y = rfr_model.predict(test_x)
```

```python
mse = mean_squared_error(test_y, pred_y)
r2_score = r2_score(test_y, pred_y)
```

```python
print(mse, r2_score)
```
